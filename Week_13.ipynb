3.	Perform linear regression on the dataset to see if you can predict water temperature based on salinity. 

```python
temp_sal_df = bottle_cleanDF[['T_degC', 'Salnty']]
X = temp_sal_df['T_degC']
y = temp_sal_df['Salnty']
```

```python
X = X.values.reshape(-1, 1)
y = y.values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 16)

reg.fit(X_train, y_train)
reg.score(X_test, y_test)
```

Salinity is not necessarily a good predictor on water temperature.

4.	Find your mean squared error and R2 values. Be sure to indicate what these tell you. You can import these from sklearn.metrics . 

```python
from sklearn.metrics import r2_score, mean_squared_error

y_pred = reg.predict(X_test)

print('r2 score ', r2_score(y_test, y_pred))
print('mean_squared_error', mean_squared_error(y_test, y_pred))
```

R-squared is the same as model score (performance).  It's a way to look at how far the data are from the regression line or polynomial. R-squared = Explained variation / Total variation

The linear model accounts for 26% of the variance in the data - not good.

MSE - mean squared error is a measure of fit and and the lower the number, the better the fit.  It tells you how the close the model line is to the data.  It is a measure of distance between the estimate and the actual point.  Squaring is used to deal with the negative values.

5.	Create a final plot of the relationship between water temperature and salinity. Be sure to show the values from the dataset in a scatter plot with a trend line that shows the predicted temperature values. 

```python
plt.figure(figsize=(10, 10))
prediction_space = np.linspace(min(X), max(X))
plt.scatter(X, y, color='blue',  edgecolors='yellow')
plt.plot(prediction_space, reg.predict(prediction_space), color='black', linewidth=3)
plt.title('Water Temp vs. Salinity, Linear Regression')
plt.xlabel('Temperature Celsius')
plt.ylabel('Salinity')

plt.show()
```

7.	Perform polynomial regression on this same dataset. Train and validate your model as you did when you used linear regression. Mean squared error and R2 both apply here. 

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3)

X_train_poly = poly.fit_transform(X_train)
poly.fit(X_train, y_train)

model = LinearRegression()
model.fit(X_train_poly, y_train)

y_pred = model.predict(poly.fit_transform(X_test))

print('r2 score ', r2_score(y_test, y_pred))
print('mean_squared_error', mean_squared_error(y_test, y_pred))
```

Using polynomail regression improved model performance from 26% to 48%.  The MSE also was reduced.
