```python
def dotProd(v1, v2):
    prod = 0
    try:
        for i in range(len(v1)):
            prod += v1[i]*v2[i]
            
    except:
        print('maybe vectors arent the same length')
    
    return prod

#not my idea, but wanted to have it somewhere
def dotProdzip(v1, v2):
    prod = 0 
    try:
        for i,n in zip(v1,v2):
            prod += i*n
    except:
        print('maybe vectors arent the same length')
    
    return prod
```

```python

a = [1,2]
b = [2, 6]
print(dotProd(a, b))
print(dotProdzip(a,b))
print(dotProdzip(north, south))
```

```python
def crossProd(v1, v2):
    c1 = 0
    c2 = 0
    c3 = 0
    try:
        c1 = v1[1]*v2[2] - v1[2]*v2[1]
        c2 = v1[2]*v2[0] - v1[0]*v2[2]
        c3 = v1[1]*v2[2] - v1[2]*v2[1]
            
    except:
        print('maybe vectors arent the same length')
    
    return np.array([c1, c2, c3])
```

```python
c = [1, 2, 3]
d = [1, 1, 1]

crossProd(c, d)
```

```python
# code room types in new column

n_list = []
for n in belize_df['neighbourhood_group']:
    if n == 'Belize':
        c = [1, 0 , 0, 0, 0, 0]
    if n == 'Stann Creek':
        c = [0, 1 , 0, 0, 0, 0]
    if n == 'Toledo':
        c = [0, 0 , 1, 0, 0, 0]
    if n == 'Cayo':
        c = [0, 0 , 0, 1, 0, 0]
    if n == 'Corozal':
        c = [0, 0 , 0, 0, 1, 0]
    if n == 'Orange Walk':
        c = [0, 0 , 0, 0, 0, 1]
    
    n_list.append(c)
    
belize_df['neighbourhood_group_array'] = n_list

belize_df.head()
```
