# DnD Classes

```python
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = 'http://dndbeyond.com/classes'
browser.visit(url)
```

```python
soup = bs(browser.html, 'html.parser')

all_classes = soup.find_all('h3', class_='listing-card__title')
```

```python
all_classes
```

```python
dnd_dict = {'class': []}
for cls in all_classes:
    # not sure .text.strip is the right way, find(xxxx).text didn't work for me
    dnd_class = cls.text.strip()
    dnd_dict['class'].append(dnd_class)
   
print(dnd_dict)
```

```python
dnd_df = pd.DataFrame(dnd_dict)
dnd_df
```
