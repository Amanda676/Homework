```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchData_df = pd.DataFrame(purchase_data)
print(purchData_df.head(5))  #see what data looks like
print(purchData_df['SN'].value_counts())  #players may make multiple purchases

playerCount = [len(purchData_df['SN'].unique())]   #number of unique players

#this code also works, saving as reference
#playerCount_df = pd.DataFrame({'Total Players' : playerCount})

playerCount_df = pd.DataFrame(playerCount, columns=["Total Players"])

playerCount_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
numUniqueItems = len(purchData_df['Item ID'].unique())
avePrice = round(purchData_df['Price'].mean(), 2)
numOfPurchases = len(purchData_df['Purchase ID'])
totalRevenue = purchData_df['Price'].sum()

purchAnalysis_df = pd.DataFrame({'Number of Unique Items': [numUniqueItems],
                                 'Average Price': ['$' + str(avePrice)], 
                                 'Number of Purchases': [numOfPurchases], 
                                 'Total Revenue':['$' + str(totalRevenue)]})

purchAnalysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
uniqueSN_df = purchData_df.drop_duplicates(subset=['SN'])
gender_df = pd.DataFrame(uniqueSN_df['Gender'].value_counts()) #I could not use columns=[], don't know why

gender_df.rename(columns={'Gender': 'Total Count'},inplace = True)  #using rename function for column
# no idea why I had to use inplace=True to get this to work

gender_df['Percentage'] = round(gender_df['Total Count']/playerCount*100,2)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
genderPurchaseAnalysis_df = pd.DataFrame(purchData_df['Gender'].value_counts())
genderPurchaseAnalysis_df.sort_index(inplace=True)
genderPurchaseAnalysis_df.rename(columns={'Gender': 'Purchase Count'},inplace = True)

genderPurchaseAnalysis_df["Avg Purchase Price"] = round(purchData_df.groupby('Gender')['Price'].mean(), 2)
genderPurchaseAnalysis_df['Avg Purchase Price'] = \
    genderPurchaseAnalysis_df['Avg Purchase Price'].map('${:,.2f}'.format)
genderPurchaseAnalysis_df["Total Purchase Value"] = purchData_df.groupby('Gender')['Price'].sum()

genderPurchaseAnalysis_df['Avg Total Purchase per Person'] = \
    round(genderPurchaseAnalysis_df['Total Purchase Value']/gender_df['Total Count'].sort_index(), 2)

genderPurchaseAnalysis_df[['Total Purchase Value', 'Avg Total Purchase per Person']] = \
    genderPurchaseAnalysis_df[['Total Purchase Value', 'Avg Total Purchase per Person']].applymap('${:,.2f}'.format)

genderPurchaseAnalysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
agesBin = [1, 9, 14, 19, 24, 29, 34, 39, 100 ]
agesLabels = ['<10','10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+' ]
purchData_df['Age Group']= pd.cut(purchData_df["Age"], agesBin, labels=agesLabels)
uniqueSN_df = purchData_df.drop_duplicates(subset=['SN']) #rerun this dataframe with new Age Group info

ageGroups = uniqueSN_df['Age Group'].value_counts().sort_index()
ageGroup_df = pd.DataFrame(ageGroups)
ageGroup_df.columns = ['Total']
ageGroup_df['Percentage of Players'] = round((ageGroup_df['Total']/playerCount)*100, 2)
ageGroup_df.index.name = 'Age Groups'

ageGroup_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
ageAnalysis_df = pd.DataFrame(purchData_df['Age Group'].value_counts().sort_index())
ageAnalysis_df.columns = ['Purchase Count']

ageAnalysis_df["Avg Purchase Price"] = round(purchData_df.groupby('Age Group')['Price'].mean(), 2)
ageAnalysis_df['Total Purchase Value'] = purchData_df.groupby('Age Group')['Price'].sum()
ageAnalysis_df['Avg Total Purchase per Person'] = ageAnalysis_df['Total Purchase Value']/ageGroup_df['Total']

ageAnalysis_df[['Avg Total Purchase per Person']] = \
    ageAnalysis_df[['Avg Total Purchase per Person']].applymap('${:,.2f}'.format)

ageAnalysis_df
```

* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order

* Display a preview of the summary data frame

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
mostPopular_df = pd.DataFrame(purchData_df[['Item ID','Item Name', 'Price']])

mostPopularGroup_df = mostPopular_df.groupby(['Item ID','Item Name'])

mostPopularResult_df = mostPopularGroup_df.count()
mostPopularResult_df.columns = ['Purchase Count']
mostPopularResult_df.sort_values(by=['Purchase Count'], ascending=False, inplace=True)

mostPopularResult_df.index.name = 'Item ID'

mostPopularResult_df['Item Price'] = mostPopularGroup_df['Price'].mean()
mostPopularResult_df['Total Purchase Value'] = mostPopularGroup_df['Price'].sum()

#running the formatting seems to mess up the next step, so commented out

#mostPopularResult_df[['Item Price', 'Total Purchase Value']] = \
#    mostPopularResult_df[['Item Price', 'Total Purchase Value']].applymap('${:,.2f}'.format)

mostPopularResult_df.head(5)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order
